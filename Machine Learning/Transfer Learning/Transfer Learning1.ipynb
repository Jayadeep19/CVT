Transfer Learning using VGG16 Pre trained Model

Using a pre trained VGG16 model on Imagenet dataset, to classify the given input image

```python
from keras.applications.vgg16 import VGG16
import tensorflow.keras.utils
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.applications.imagenet_utils import decode_predictions
```

```python
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
```

```python
model = VGG16()
```

```python
model.summary()
```

```python
image1 = tensorflow.keras.utils.load_img('/content/drive/MyDrive/datasets/dog.2175.jpg', target_size = (224,224))
image1 = tensorflow.keras.utils.img_to_array(image1)
image1 = image1.reshape((1, image1.shape[0], image1.shape[1], image1.shape[2]))
image1 = preprocess_input(image1)
```

```python
predict1 = model.predict(image1)  
class_predicted = decode_predictions(predict1, top =5)
print(class_predicted)
print(len(class_predicted))
```

```python
label = class_predicted[0][1]
print(label)
print(label[1])
```

```python
list = []
for i in range (4):
  print('The top {} classification result is:'.format(i+1))
  print('{0} with probability of {1}%'.format(class_predicted[0][i][1], class_predicted[0][i][2]*100))
  
```
