```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
hsv = cv2.imread('HSV Values.png')
plt.imshow(hsv)
```

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lower = np.array([95,50,20])
    higher = np.array([105,255,250])
    
    mask = cv2.inRange(hsv, lower, higher)
    
    contors,_ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
    
    if contors:
        cv2.drawContours(frame,contors, -1,(0,255,0), 2)
    
    
    
    cv2.imshow("window", frame)
    cv2.imshow("window1", hsv)
    cv2.imshow("window2", mask)
    
    k = cv2.waitKey(1)
    
    if k == ord('s'):
        cv2.imwrite('contour.jpeg',frame)
        cv2.imwrite('contour1.jpeg',hsv)
        cv2.imwrite('contour2.jpeg',mask)
        pic1 = cv2.imread('contour.jpeg')
        pic2 = cv2.imread('contour1.jpeg')
        pic3 = cv2.imread('contour2.jpeg')
        f,arr = plt.subplots(1,3)
        arr[0].imshow(pic1)
        arr[1].imshow(pic2)
        arr[2].imshow(pic3)
    
    if k ==27:
        cap.release()
        cv2.destroyAllWindows()
        break
```
